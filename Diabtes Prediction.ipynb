```python
#seperating data and variables
X=df.drop(columns="Outcome",axis=1)
Y=df["Outcome"]
```

```python
print(X,Y)
```

```python
#Standardize the Data using StandardScalar
scalar = StandardScaler()
scalar.fit(X)
```

```python
standard_data = scalar.transform(X)
```

```python
standard_data
```

```python
X=standard_data
Y=df["Outcome"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=.20, random_state=1,stratify=Y)
```

```python
print(x_test,x_train,y_test,y_train)
```

```python
Classifier = svm.SVC(kernel='linear')
```

```python
#Training the support vector machine classifier
Classifier.fit(x_train,y_train)
```

Model Evaluation

```python
#ACCURACY SCORE
x_train_accuracy = Classifier.predict(x_train)
training_accuracy = accuracy_score(x_train_accuracy,y_train)
print(x_train_accuracy)
print(training_accuracy)
```

```python
print("Accuracy Score on our Training data", training_accuracy)
```

```python
#ACCURACY SCORE  TEST DATA
x_test_accuracy = Classifier.predict(x_test)
test_accuracy = accuracy_score(x_test_accuracy,y_test)
# print(x_test_accuracy)
print("Test data accuracy",test_accuracy)
```

Making Prediction

```python
input_data = (9,171,110,24,240,45.4,0.721,54)
#changing this data into numpy array
input_data_Ary = np.asarray(input_data).reshape(1,-1)
#standarsize the input data
std_data = scalar.transform(input_data_Ary)
print("standard_data",std_data)
prediction = Classifier.predict(std_data)
# print(prediction)
if prediction[0]==0:
    print("The Person has not diabetics")
else:
    print("The person has diabetics")
```
